```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('mdb184_output.png',-1)
#img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
max_pixVal = np.amax(img)


_, th1 = cv2.threshold(img, max_pixVal*0.91, 255, cv2.THRESH_TOZERO) #cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2) 

#cv2.imshow("Image", img)
#cv2.imshow("THRESH_BINARY", th1)
#cv2.imshow("ADAPTIVE_THRESH_MEAN_C", th2)
#cv2.imshow("ADAPTIVE_THRESH_GAUSSIAN_C", th3)

#cv2.waitKey(0)
#cv2.destroyAllWindows()



img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
th1 = cv2.cvtColor(th1, cv2.COLOR_GRAY2BGR)
th2 = cv2.cvtColor(th2, cv2.COLOR_GRAY2BGR)
th3 = cv2.cvtColor(th3, cv2.COLOR_GRAY2BGR)




fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharey=True)

axs[0, 0].set_title('Original Image')
axs[0, 0].imshow(img)

axs[0, 1].set_title('Thresh Binary')
axs[0, 1].imshow(th1)

axs[1, 0].set_title('Adaptive Thresh Mean C')
axs[1, 0].imshow(th2)

axs[1, 1].set_title('Adaptive Thresh Gaussian C')
axs[1, 1].imshow(th3)

fig.suptitle('ADAPTIVE THRESHOLDING')

plt.show()
```
